```python
## Chat completion API


## COMPLETION API

response = openai.Completion.create(
    model = 'text-search-curie-query-001',
    prompt ='Act like an AI assistant and explain machine learning to a 5 year old kid',
    max_tokens = 400,
    temperature = 0
)
print(response)
```

```python
print(response["choices"][0]['text'])
```

```python
## Chat completion API

response = openai.Completion.create(
    model = 'text-davinci-003',
    prompt ='Act like an AI assistant and explain machine learning to a 5 year old kid',
    max_tokens = 20,
    temperature = 1,
    n=5
)
print(response)
```

```python
print(response["choices"])
```

```python
## Text completion or text generation

response = openai.Completion.create(
    model = 'text-davinci-003',
    prompt ='What is the height of Mount Everest',
    max_tokens = 20,
    temperature = 1,
    n=5
)
print(response)
```
